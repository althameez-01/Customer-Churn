The Chi-Squared () test is a statistical test used to determine whether there is a significant association between two categorical variables  typically one feature and the target (class label).

```python
# prompt: statistical test to identify feature importance for classification of churn

import pandas as pd
from scipy.stats import chi2_contingency

# Identify categorical features from X_cat (excluding the target if it was still there)
categorical_features = X_cat.columns

# Perform Chi-Square test for independence between each categorical feature and the target variable 'Churn'
print("Chi-Square Test Results for Categorical Feature Importance:")
print("-" * 50)

for feature in categorical_features:
    # Create a contingency table
    contingency_table = pd.crosstab(X_cat[feature], y)

    # Perform the Chi-Square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Print the results
    print(f"Feature: {feature}")
    print(f"  Chi-Square Statistic: {chi2:.4f}")
    print(f"  P-value: {p:.4f}")

    # Interpret the p-value (common significance level is 0.05)
    if p < 0.05:
        print("  Interpretation: Reject the null hypothesis. There is a significant association between the feature and Churn.")
    else:
        print("  Interpretation: Fail to reject the null hypothesis. There is no significant association between the feature and Churn.")

    print("-" * 50)

# For numerical features, while not a statistical test of "importance" in the same way as chi-square for categorical,
# you can look at correlation with the target variable (as done in the preceding code)
# or use other methods like t-tests if the assumptions are met.
# Since correlation was already computed, we can reiterate the findings or look at the absolute correlation values.

print("\nCorrelation with Churn (for Numerical Features):")
print("-" * 50)

# Ensure y is numeric for correlation calculation
y_numeric = y_encoded # Using the already encoded target

# Create a dataframe with numerical features and the numeric target
numerical_df_with_target = pd.concat([X_num, pd.Series(y_numeric, name='Churn')], axis=1)

# Calculate the correlation of numerical features with Churn
correlation_with_churn = numerical_df_with_target.corr()['Churn'].drop('Churn')

# Print the correlations
print(correlation_with_churn)
print("-" * 50)

# Interpretation: Higher absolute correlation values indicate a stronger linear relationship with churn.

```

```python
# prompt: In this statistical test? what is the signicance level and cofidence level

In the context of the Chi-Square test performed in the preceding code:

*   **Significance Level (Alpha):** The significance level is not explicitly defined as a variable in the provided code, but it is implied and used in the interpretation. The common significance level used for hypothesis testing, and which the code uses for interpretation (`if p < 0.05:`), is **0.05**. This means that if the p-value is less than 0.05, we reject the null hypothesis (that there is no association between the feature and Churn). A significance level of 0.05 indicates a 5% risk of concluding that an association exists when there is no actual association (Type I error).

*   **Confidence Level:** The concept of a confidence level is related to confidence intervals, which are not calculated or used in the provided Chi-Square test code. The Chi-Square test primarily provides a p-value to assess the statistical significance of an association. While a significance level of 0.05 is often associated with a 95% confidence level in other statistical contexts (like confidence intervals), the Chi-Square test itself doesn't directly involve calculating a confidence level for the association. Therefore, there is **no confidence level explicitly stated or used** in the provided Chi-Square test code.

For the correlation analysis on numerical features, again, there is **no confidence level explicitly stated or used**. Correlation calculates the strength and direction of a linear relationship. While confidence intervals for correlation coefficients can be calculated, they are not in this code snippet.

In summary:
*   **Significance Level:** 0.05 (implied and used for p-value interpretation)
*   **Confidence Level:** Not applicable/Not used in this code.
```

```python
# prompt: Then on what basis chi squred test statistic results displayed

import pandas as pd
In the provided Python code snippet, the Chi-Square test statistic results are displayed based on the following:

1.  **Contingency Table:** For each categorical feature, a contingency table (`contingency_table = pd.crosstab(X_cat[feature], y)`) is created. This table shows the frequency counts of the combinations of categories for the feature and the target variable (`y`).

2.  **`chi2_contingency()` function:** The `scipy.stats.chi2_contingency()` function is then applied to this contingency table. This function calculates several values, including the Chi-Square test statistic (`chi2`), the p-value (`p`), the degrees of freedom (`dof`), and the expected frequencies (`expected`).

3.  **Chi-Square Statistic:** The `chi2` value returned by the function is the Chi-Square test statistic. This statistic is a measure of the difference between the observed frequencies in the contingency table and the expected frequencies (the frequencies that would be expected if there were no association between the two variables). A larger Chi-Square statistic indicates a greater difference between observed and expected frequencies, suggesting a stronger association.

4.  **P-value:** The `p` value returned is the p-value associated with the calculated Chi-Square statistic. The p-value represents the probability of observing a Chi-Square statistic as extreme as, or more extreme than, the one calculated from the data, assuming that the null hypothesis (no association between the variables) is true.

5.  **Degrees of Freedom:** The `dof` value represents the degrees of freedom for the Chi-Square test, which is calculated based on the dimensions of the contingency table.

6.  **Interpretation:** The code then uses the p-value to interpret the results based on a significance level of 0.05.
    *   If the p-value (`p`) is less than 0.05, the code concludes that there is a statistically significant association between the feature and the target variable ('Churn'). The null hypothesis of independence is rejected.
    *   If the p-value is greater than or equal to 0.05, the code concludes that there is no statistically significant association at the 0.05 level. The null hypothesis of independence is failed to be rejected.

Therefore, the Chi-Square test statistic results are displayed *after* the `chi2_contingency` function calculates the statistic based on the comparison of observed frequencies in the contingency table to the expected frequencies under the assumption of independence. The interpretation of these results (specifically, the p-value) then determines whether a significant association is reported.
```

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer

# List of categorical features
label_cat = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines',
             'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
             'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling']

contract_order = [['Month-to-month', 'One year', 'Two year']]
ordinal_cat = ['Contract']
nominal_cat = ['InternetService', 'PaymentMethod']

# FunctionTransformer wrapper for Label Encoding binary categorical variables
def binary_label_encode(df):
    return df.apply(lambda col: col.map({'Yes': 1, 'No': 0}))

label_encoder_transformer = Pipeline(steps=[
    ('label_encode', FunctionTransformer(binary_label_encode, validate=False))
])

# Ordinal Encoder for 'Contract'
ordinal_encoder = Pipeline(steps=[
    ('ordinal_encode', OrdinalEncoder(categories=contract_order))
])

# One-Hot Encoder for Nominal Categorical
nominal_encoder = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine all in ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    ('label_enc', label_encoder_transformer, label_cat),
    ('ordinal_enc', ordinal_encoder, ordinal_cat),
    ('onehot_enc', nominal_encoder, nominal_cat)
])

# Example usage: fit_transform on your categorical features
# Assuming X_cat is your categorical feature DataFrame with all necessary columns
# X_cat should include: label_cat + ordinal_cat + nominal_cat

# transformed_cat = preprocessor.fit_transform(X_cat)

# Optional: convert to DataFrame with readable column names (for inspection)
# transformed_df = pd.DataFrame(transformed_cat)

```
